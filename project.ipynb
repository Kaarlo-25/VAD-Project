# NYC TAXIS VISUALIZATION

```python
import pandas as pd
import folium
import geopandas as gpd
import geodatasets
```

```python
data_1 = pd.read_csv("C:\\Users\\valko\\OneDrive\\Documentos\\data\\yellow_tripdata_2016-01.csv")
data_1
```

```python
data_sample = data_1.sample(n=1000, random_state=42)  # `random_state` asegura que siempre obtengas la misma muestra
data_sample.head()
```

```python
nyc_map = gpd.read_file(geodatasets.get_path("geoda.nyc"))

# Asignar CRS si no estÃ¡ definido
if nyc_map.crs is None:
    nyc_map.set_crs(epsg=4326, inplace=True)

# Crear el mapa centrado en NYC
m = folium.Map(
    location=[40.7128, -74.0060],
    zoom_start=11,
    tiles="CartoDB Positron",
    min_zoom=10,
)

for _, row in data_sample.iterrows():
    folium.CircleMarker(
        location=[row["pickup_latitude"], row["pickup_longitude"]],
        radius=3,  # TamaÃ±o del punto
        color="green",
        fill=True,
        fill_color="green",
        fill_opacity=0.7
    ).add_to(m)

m  # Mostrar el mapa
```
